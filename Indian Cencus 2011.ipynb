# Data Visualisation

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Population demographics

```python
state_population = df.groupby('State_name')['Population'].sum()
state_population = state_population.sort_values(ascending=False)

sns.set(rc={'figure.figsize':(20,10)})
state_population.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Population')
plt.title('State-wise Population')
plt.show()
```

### Religious configuration

```python
state_religious_data = df.groupby('State_name' , as_index=False)[['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']].sum()
state_religious_data
```

```python
sns.set(rc={'figure.figsize':(20,10)})
state_religious_data.plot(kind='bar', stacked=True)
plt.xlabel('State')
plt.ylabel('Population')
plt.title('Religious Composition by State')
plt.legend(title='Religion')
plt.show()
```

#### For individual state 

```python
def religious_demo_of_state(df, state):
    state_data = df[df['State_name'] == state]
    religion_names = state_data.columns[1:]
    population_values = state_data.values.flatten()[1:]
    
    # Pie chart
    fig, ax = plt.subplots()
    colors = sns.color_palette('cool')
    ax.pie(population_values, labels=religion_names, autopct='%1.1f%%',colors=colors)
    ax.set_title('Religion Distribution in {}'.format(state))
    plt.show()
```

```python
religious_demo_of_state(state_religious_data, 'KERALA')
```

### Age distribution 

```python
plt.scatter(df['State_name'], df['Age_Group_0_29'], label='Age Group 0-29')
plt.scatter(df['State_name'], df['Age_Group_30_49'], label='Age Group 30-49')
plt.scatter(df['State_name'], df['Age_Group_50'], label='Age Group 50+')

plt.xlabel('State')
plt.ylabel('Population')
plt.title('Age Composition of the State')
plt.legend(title='Age Group')
plt.xticks(rotation=90) 
plt.show()
```

#### For Individual State

```python
state_age_data = df.groupby('State_name' , as_index=False)[['Age_Group_0_29', 'Age_Group_30_49', 'Age_Group_50']].sum()
```

```python
def generate_bar_plot(df, state):
    state_data = df[df['State_name'] == state]
    age_ = state_data.columns[1:]
    population_values = state_data.values.flatten()[1:]
    
    colors = sns.color_palette('cool')
    sns.set_style('whitegrid')
    ax = sns.barplot(x=population_values, y=age_, palette=colors)
    ax.set_title('Age distribution in {}'.format(state), fontweight='bold')
    plt.show()
```

```python
generate_bar_plot(state_age_data, 'KERALA')
```

### Gender Wise visualization

```python
gender_wise_data = df.groupby('State_name' , as_index=False)[['Population','Male','Female']].sum()
gender_wise_data
```

```python
gender_ratio = []
for i in range(len(gender_wise_data['State_name'])):
    male_population_state = gender_wise_data.iloc[i, 1:].iloc[1]
    female_population_state = gender_wise_data.iloc[i, 1:].iloc[2]
    ratio = male_population_state / female_population_state
    gender_ratio.append(ratio)
```

```python
gender_wise_data['Gender_ratio'] = gender_ratio
gender_wise_data
```

```python
female_pop = gender_wise_data['Female'].sum()
female_pop_avg = female_pop/35

print("Total female population is ",female_pop)
print("Below are the states who have women more than the avarage")
print(gender_wise_data[gender_wise_data['Female'] > female_pop_avg]['State_name'])
print("These states have more number of women")
gender_wise_data[gender_wise_data['Gender_ratio'] < 1]['State_name']
```

```python
female_pop = gender_wise_data['Female'].sum()
female_pop_avg = female_pop/35
filtered_states = gender_wise_data[gender_wise_data['Female'] > female_pop_avg]['State_name']
gender_imbalance_states = gender_wise_data[gender_wise_data['Gender_ratio'] < 1]['State_name']

print(f"# Total female population: {female_pop}")
print("# Below are the states with a female population above the average:")
print(filtered_states.to_string(index=False))
print("# These states have a gender imbalance with more women:")
print(gender_imbalance_states.to_string(index=False))
```

```python
sns.barplot(x='State_name', y='Gender_ratio', data=gender_wise_data)
plt.xlabel('State')
plt.ylabel('Gender Ratio')
plt.title('Gender Ratio by State')
plt.xticks(rotation=90) 
plt.show()
```

## Occupation wise

```python
occupation_wise_data = df.groupby('State_name' , as_index=False)[['Workers', 'Male_Workers', 'Female_Workers','Cultivator_Workers', 'Agricultural_Workers', 'Household_Workers']].sum()
occupation_wise_data
```

```python
sns.set(rc={'figure.figsize':(25,10)})
occupation_wise_data[['Male_Workers','Female_Workers']].plot(kind='bar' ,stacked=False)
plt.xlabel('State')
plt.ylabel('Population')
plt.title('Population in workspace by the Male and the Female ')
plt.legend(title='Age')
plt.show()
```

```python
def state_wise_pop_workclass(df, state):
    state_data = df[df['State_name'] == state]
    Job_class = state_data.columns[4:]
    population_values = state_data.values.flatten()[4:]
    
    # Pie chart
    fig, ax = plt.subplots()
    colors = sns.color_palette('cool')
    ax.pie(population_values, labels=Job_class, autopct='%1.1f%%',colors=colors)
    ax.set_title('Workclass Distribution in {}'.format(state))
    plt.show()
```

```python
state_wise_pop_workclass(occupation_wise_data, "BIHAR")
```
